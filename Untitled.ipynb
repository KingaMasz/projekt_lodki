```python
def load_titanic_data(filename, titanic_path=TITANIC_PATH):
    csv_path = os.path.join(titanic_path, filename)
    return pd.read_csv(csv_path, encoding="utf8", encoding_errors='ignore', on_bad_lines='skip')
```
